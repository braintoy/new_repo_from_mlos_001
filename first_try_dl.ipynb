```python
#https://scikit-learn.org/stable/modules/classes.html#module-sklearn.metrics
#pip3 install sklearn
# df is our dataframe
# 1. Load data 
# 2. create input and target list
# 3 cross-validation Dataset
# 4. Modeling 
# 1. Load data 
import pandas as pd
import pickle 
# Way one
# from sklearn import model_selection
# model_selection.train_test_split()
from sklearn.model_selection import train_test_split


df=pd.read_csv("irish.csv",encoding="UTF-8",low_memory=False)
# 2. create input and target list
inpfeatures=['sepal-length', 'sepal-width', 'petal-length', 'petal-width']
target=  ['species']
X = df[inpfeatures]
y= df[target]
ts=0.2
# 3. cross-validation Dataset 80/20 split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=ts)


# fn="tst_data.pkl"
# with open(fn,'wb') as writebuf:
#     pickle.dump(X_test, writebuf)
# fn="tst_lbl.pkl"
# with open(fn,'wb') as writebuf:
#     pickle.dump(y_test, writebuf)

    
    
# from sklearn.neighbors import  KNeighborsClassifier
# from sklearn.svm import LinearSVR 
# model = LinearSVR()
# model.fit(X_train,y_train.values.ravel())
    



```

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.utils import to_categorical
# https://numpy.org/doc/stable/reference/generated/numpy.unique.html
# print(y_train)
# print(df["species"].to_numpy)
def my_ohe(col_as_df_col):
    myarray=col_as_df_col.to_numpy()
    uniques, ids= np.unique(myarray,return_inverse=True)
    return to_categorical(ids, len(uniques))

train_y_ohe= my_ohe(y_train)
print(train_y_ohe)

test_y_ohe= my_ohe(y_test)

print(train_y_ohe)

print(test_y_ohe)

# print(train_y_ohe)

# a, b = train_y_ohe.shape

# print(a)
# print(b)
```

```python
# model = tf.keras.models.Sequential(
#     [
#   tf.keras.layers.Flatten(input_shape=(4, 1)),
#   tf.keras.layers.Dense(3, activation='relu'),
#   tf.keras.layers.Dense(6, activation='relu'),
#   tf.keras.layers.Dense(4, activation='relu'),
#   tf.keras.layers.Dropout(0.2),
#   tf.keras.layers.Dense(1)
# ]
# )

model = tf.keras.Sequential([
  tf.keras.layers.Dense(10, activation=tf.nn.relu, input_shape=(4,)),  # input shape required
  tf.keras.layers.Dense(10, activation=tf.nn.relu),
  tf.keras.layers.Dense(3)
])

model.summary()

model.compile(optimizer='adam',loss=tf.keras.losses.BinaryCrossentropy(from_logits=True), metrics=['accuracy'])


# xTrn=tf.convert_to_tensor(X_train.values)

# print(xTrn)

# # print(X_train)
# model.compile(optimizer='adam', metrics=['accuracy'])

# model.fit( xTrn, train_y_ohe, epochs=1)
```
